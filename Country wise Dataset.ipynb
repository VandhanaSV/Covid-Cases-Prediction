


# Covid-19 Dataset

Attribute Imformation

1. *Country/Region:* The name of the country or region for which the COVID-19 data is reported.

2. *Confirmed:* The total number of confirmed COVID-19 cases in the specified country or region.

3. *Deaths:* The total number of deaths attributed to COVID-19 in the specified country or region.

4. *Recovered:* The total number of individuals who have recovered from COVID-19 in the specified country or region.

5. *Active:* The total number of currently active COVID-19 cases (Confirmed - Deaths - Recovered).

6. *New cases:* The number of new confirmed COVID-19 cases reported.

7. *New deaths:* The number of new deaths attributed to COVID-19 reported.

8. *New recovered:* The number of new recoveries from COVID-19 reported.

9. *Deaths / 100 Cases:* The percentage of confirmed cases that resulted in death (Deaths / Confirmed * 100).

10. *Recovered / 100 Cases:* The percentage of confirmed cases that have recovered (Recovered / Confirmed * 100).

11. *Deaths / 100 Recovered:* The percentage of recovered cases that resulted in death (Deaths / Recovered * 100).

12. *Confirmed last week:* The total number of confirmed COVID-19 cases reported in the previous week.

13. *1 week change:* The change in the total number of confirmed cases compared to the previous week (Confirmed - Confirmed last week).

14. *1 week % increase:* The percentage increase in the total number of confirmed cases compared to the previous week (1 week change / Confirmed last week * 100).

15. *WHO Region:* The World Health Organization (WHO) region to which the country or region belongs. WHO regions are geographic regions used for administrative purposes.

    These columns provide a comprehensive overview of the COVID-19 situation in different countries or regions, including the number of confirmed cases, deaths, recoveries, and various metrics that help analyze the impact and progression of the pandemic.
